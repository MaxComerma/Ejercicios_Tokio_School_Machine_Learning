```python
# Cargamos las librerias
import requests
from urllib.request import urlretrieve
import os
```

```python
PREDICTION_KEY = ""
PREDICTION_URL = "https://max23.cognitiveservices.azure.com/customvision/v3.0/Prediction/d295522f-38db-47d6-b33f-6c92590c8597/classify/iterations/Iteration1/url"

test_images = {
    "Perro 1": "https://upload.wikimedia.org/wikipedia/commons/thumb/2/26/YellowLabradorLooking_new.jpg/1200px-YellowLabradorLooking_new.jpg",
    "Perro 2": "https://upload.wikimedia.org/wikipedia/commons/thumb/1/18/Dog_Breeds.jpg/1200px-Dog_Breeds.jpg",
    "Gato 1": "https://upload.wikimedia.org/wikipedia/commons/thumb/4/4d/Cat_November_2010-1a.jpg/1200px-Cat_November_2010-1a.jpg",
    "Gato 2": "https://upload.wikimedia.org/wikipedia/commons/thumb/b/bb/Kittyply_edit1.jpg/1200px-Kittyply_edit1.jpg",
}

headers = {
    "Prediction-Key": PREDICTION_KEY,
    "Content-Type": "application/json"}

```

```python
for nombre, url in test_images.items():
    body = {"Url": url}
    response = requests.post(PREDICTION_URL, headers=headers, json=body)
    preds = response.json()["predictions"]
    resultado = max(preds, key=lambda x: x["probability"])
    print(f"{nombre} → {resultado['tagName']}: {resultado['probability']*100:.2f}%")
```

El modelo ha conseguido un 100% de precisión y recall en ambas clases. No se ha equivocado ni una sola vez clasificando perros y gatos.

Para el entrenamiento usamos un total de 88 imágenes, 44 de perro y 44 de gato.

Dicho esto, el modelo tiene limitaciones claras. Está entrenado únicamente con perros y gatos, así que si le pasamos una imagen de cualquier otro animal, intentará clasificarlo igualmente como perro o gato aunque no lo sea.
